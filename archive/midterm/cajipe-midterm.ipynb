### Change the original DataFrame using the `inplace = True` argument to remove all rows with NULL values

**Note:** Now, the dropna(inplace = True) will NOT return a new DataFrame, but it will remove all rows containing NULL values from the original DataFrame.

## Removing Duplicates

*Duplicate rows* are rows that have been registered **more than one time**.
By taking a look at our test data set, we can assume that the `id` column is a unique identifier for each row, and thus, each `id` should only appear once.

To discover duplicates, we can use the `duplicated()` method.

The `duplicated()` method returns a Boolean values for each row:

```python
df
```

### Check for Duplicates

```python
# Returns True for every row that is a duplicate, otherwise False:
df.duplicated()
```

Question: What happened to the old data set? Explain your answer.

**Answer: It showed boolean results for every index.**

### Remove Duplicates in `id` column


To remove duplicates, use the drop_duplicates() method.

```python
# Remove duplicates in UserID:
df.drop_duplicates(subset = 'UserID', keep = 'first', inplace = True)
```

**Note:** The `(inplace = True)` will make sure that the method does NOT return a new DataFrame, but it will remove all duplicates from the original DataFrame.


### Check if there are duplicates again

```python
df
```

```python
df.shape
```

```python
df.duplicated()
```

Question: What happened to the old data set? Explain your answer.


**Answer: We have removed the duplicates, and the `duplicated()` method now returns `False` for every row.**

# 2. DATA CORRELATION

Identify the good and bad relationships in the given data set.


## FINDING GOOD RELATIONSHIPS

### A. To show relationship between columns

```python
# For showing relationship of all the columns

df.corr()
```

```python
# For showing relationship between two specific columns
df["Duration"].corr(df["Calories"])
```

### B. Add a parameter method in corr() function

```python
# pearson: standard correlation coefficient 
df.corr('pearson')
```

```python
# spearman: Spearman rank correlation
df.corr('spearman')
```

```python
# kendall: Kendall Tau correlation coefficient 
df.corr('kendall')
```

### 1. Find and identify data with good correlation.

Using Pearson correlation as default

```python
# Find and identify data with good correlation.

# Use any parameter method to find good correlation
df.corr(method='pearson')
```

**What are the data with good correlation?**

```python
# Show relationship between two or more specific columns that show good correlation
df["Duration"].corr(df["Calories"])
```

**Duration and Calories Burned have a good correlation because the value is close to 1.**

### 2. Find and identify data with bad correlation.


Using Pearson correlation as default

```python
df.corr(method='pearson')
```

**What are the data with bad correlation?**

```python
# Show relationship between two or more specific columns that show bad correlation
df["Duration"].corr(df["Maxpulse"])
```
