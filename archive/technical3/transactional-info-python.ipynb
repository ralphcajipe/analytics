Formulate 10 analytical questions(reports) that can be answered based on the columns of the dataset.

1. How many customers are there?
2. How many customers that have not been active for 6 months?
3. What is the average tenure of the customers?
4. How many customers have a deposit?
5. How many customers have a loan?
6. How many customers are active?
7. What is the average deposit balance?
8. What is the average loan balance?
9. What is the average amount of transaction?
10. What is the average amount of loan transaction?

Answer the questions using the data.

```python
# 1. How many customers are there?
df['CUSID'].nunique()
```

```python
# 2. How many customers that have not been active for 6 months?
df['CUSID'].nunique() - df[df['MONTHS_LAST_ACTIVE'] < 6]['CUSID'].nunique()
```

```python
# 3. What is the average tenure of the customers?
df['TENURE_MONTHS'].mean()
```

```python
# 4. How many customers have a deposit?
df[df['DEPOSIT INCOME'] > 0]['CUSID'].nunique()
```

```python
# 5. How many customers have a loan?
df[df['LOAN INCOME'] > 0]['CUSID'].nunique()
```

```python
# 6. How many customers are active?
df[df['MONTHS_LAST_ACTIVE'] < 6]['CUSID'].nunique()
```

```python
# 7. What is the average deposit balance?
df['DEPOSIT_BALANCE'].mean()
```

```python
# 8. What is the average loan balance?
df['LOAN_BALANCE'].mean()
```

```python
# 9. What is the average amount of transaction?
df['AMOUNT_TRANSACTION_CURRENT'].mean()
```

```python
# 10. What is the average amount of loan transaction?
df[df['LOAN INCOME'] > 0]['AMOUNT_TRANSACTION_CURRENT'].mean()
```

### 4.5 What are the anomalies present on the given data?

```python
# There are two TENURE_MONTHS columns that have the same name and the same values, so we can drop
# the one with .1 at the end
df.drop('TENURE_MONTHS.1', axis=1, inplace=True)
```

```python
# There are two AGE_MONTHS columns that have the same name and the same values, so we can drop
# the one with .1 at the end
df.drop('AGE_MONTHS.1', axis=1, inplace=True)
```
